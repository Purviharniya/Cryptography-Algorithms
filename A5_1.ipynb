```python
def maj(x8,y10,z10):
    if(x8 + y10 + z10 > 1):
        return 1
    else:
        return 0
    
def gen_keystream(X,Y,Z):
    keystream=[]
    for i in range(1,33):
        
        M=maj(X[8],Y[10],Z[10])

        if X[8]==M:
            tx=X[13]^X[16]^X[17]^X[18]
            X.pop()
            X.insert(0,tx)

        if Y[10]==M:
            ty=Y[20]^Y[21]
            Y.pop()
            Y.insert(0,ty)

        if Z[10]==M:
            tz=Z[7]^Z[20]^Z[21]^Z[22]
            Z.pop()
            Z.insert(0,tz)

        kbit=X[18]^Y[21]^Z[22]    
        keystream.append(kbit)
        print("\n--------Iteration {}--------".format(i))
        print("X= ",("".join([str(i) for i in X])))
        print("Y= ",("".join([str(i) for i in Y])))
        print("Z= ",("".join([str(i) for i in Z])))
        print("M= ",M)
        print("Keystream bit= ",kbit)
    
    print("\nKeystream Generated:",("".join([str(i) for i in keystream])))
    return keystream    

def to_binary(pt):
    s = ""
    i = 0
    for i in pt:
        binary = str(' '.join(format(ord(x), 'b') for x in i))
        j = len(binary)
        while(j < 8):
            binary = "0" + binary
            s = s + binary
            j = j + 1
    binary_values = []
    k = 0
    while(k < len(s)):
        binary_values.insert(k, int(s[k]))
        k = k + 1
    return binary_values

def to_str(cipher):
    s = ""
    length = len(cipher) - 8
    i = 0
    while(i <= length):
        t=chr(int(cipher[i:i+8], 2))
        s = s + t
        i = i + 8
    return str(s)

def get_key(pt,ks):
    if(len(pt)>len(ks)):
        bits_to_append= len(pt)-32
        for i in range(bits_to_append):
            ks.append(ks[i])
            
    if(len(pt)<len(ks)):
        ks=ks[0:len(pt)]
        
    return ks

def encrypt(pt,ks):
    pt=to_binary(pt)
    key_li=get_key(pt,ks)
    bin_cipher=""
    for i in range(len(pt)):
        bin_cipher+= str(pt[i]^key_li[i])
        
    cipher=to_str(bin_cipher)
    print("Cipher in binary format: ",bin_cipher)
    print("Cipher in String format: ",cipher)

def decrypt(cipher,ks):
    cipher=[int(i) for i in cipher]
    key_li=get_key(cipher,ks)
    pt=""
    for i in range(len(cipher)):
        pt+=str(cipher[i]^key_li[i])

    pt_str=to_str(pt)
    print("Plaintext in binary format:",pt)
    print("Plaintext in string format:",pt_str)   

while(True):
    ch=int(input("1.Encryption\n2.Decryption\n3.Exit\nEnter here:"))
    if(ch==3):
        break
        
    elif(ch==1):
        X=input("Enter the 19 bits in X register:")
        Xbits=[int(i) for i in X]
        Y=input("Enter the 22 bits in Y register:")
        Ybits=[int(i) for i in Y]
        Z=input("Enter the 23 bits in Z register:")
        Zbits=[int(i) for i in Z]
        plain=input("Enter the plaintext:")
        keystream= gen_keystream(Xbits,Ybits,Zbits)   
        encrypt(plain,keystream)
        
    elif(ch==2):
        X=input("Enter the 19 bits in X register:")
        Xbits=[int(i) for i in X]
        Y=input("Enter the 22 bits in Y register:")
        Ybits=[int(i) for i in Y]
        Z=input("Enter the 23 bits in Z register:")
        Zbits=[int(i) for i in Z]
        cipher=input("Enter the ciphertext in binary format:")
        keystream= gen_keystream(Xbits,Ybits,Zbits)   
        decrypt(cipher,keystream)
    
#1010101010101010101
#1100110011001100110011
#11100001111000011110000
#1010101010101010101110011001100110011001111100001111000011110000
```
