```python
import math

def encrypt(keys,plaintext):
    plaintext=plaintext.replace(' ', '') 
    indices=[i for i in range(len(keys))]
    dic=dict(zip(keys,indices))
    so=sorted(dic.keys())
    col=len(keys)                                
    row = int(math.ceil( len(plaintext)/ col))
    bogus="x"*((row*col)-len(plaintext))
    plaintext+=bogus  #adding required bogus characters

    equalStr=[]
    for i in range(0, len(plaintext), col):  #Dividing string in parts equal to rows
            part = plaintext[ i : i+col];  
            equalStr.append(part);

    cipher=""
    for part in equalStr:
        for i in so:
            cipher+=part[dic[i]]
            
    return cipher

def decrypt(keys,cipher):
    col=len(keys)
    row = int(math.ceil( len(cipher)/ col))
    listi=[i for i in range(col)]
    dic=dict(zip(listi,keys))
    sorted_keys=sorted(keys)
    dic2=dict(zip(listi,sorted_keys))

    cipher=cipher.replace(' ', '') 

    str=''
    for i in listi:
        str+=cipher[i:len(cipher):col]  #arranging the cipher text columnwise   

    equalStr=[]
    for i in range(0, len(cipher), row):  #Dividing string in parts equal to cols, each part having chars equal to rows
            part = str[ i : i+row];  
            equalStr.append(part);
#     print(equalStr)

    cipher_pos=[]
    for i in list(dic2.values()): 
        for key,val in list(dic.items()):
            if i==val and key not in cipher_pos:
                cipher_pos.append(key)    #temporary list to store indices of key items columnwise

    pt_li=[]
    for i in range(col):
        if i in cipher_pos:
            index=cipher_pos.index(i)
            pt_li.append(equalStr[index])

    pt=''

    for temp in range(row):
        for i in pt_li:
            pt+=i[temp]
    return pt

while True:
    ch=input("\nEnter the option:\n1.Encryption\n2.Decryption\n3.Exit\nEnter here:")
    if ch=='1':
        plaintext=input("\nEnter the Plain text:")
        key=[int(i) for i in input("Enter the keys:").split()]
        print("\nThe encrypted message is:",encrypt(key,plaintext))
    
    elif ch=='2':
        cipher=input("\nEnter the Cipher text:")
        key=[int(i) for i in input("Enter the keys:").split()]
        print("\nThe decrypted message is:",decrypt(key,cipher))

    elif ch=='3':
        print("\nProgram Terminated!")
        break
        
    else:
        print("\nInvalid choice! enter again")
#the simplest possible transposition
#2 5 4 1 3
```
