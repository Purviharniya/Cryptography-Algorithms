```python
import numpy as np

def gen_key(N,key):
    k1=0
    k2=N
    keyli=[]
    for i in range(len(key)//N):
        keyli.append(int(key[k1:k2],2))
        k1+=N
        k2+=N
    return keyli

def convert_str_to_bin(N,pt):
    h=''
    for i in pt:
        h+=(hex(ord(i))[2:])
        
    scale = 16 ## equals to hexadecimal
    f='0'
    f+=bin(int(h, scale))[2:].zfill(8)
    t1=0
    t2=N
    plainstr=[]
    for i in range(2**N):
        plainstr.append(int(f[t1:t2],2))
        t1+=N
        t2+=N
    return plainstr

def convert_to_str(msg_li,N):
    li=list(msg_li)
    t=''
    inc=2**N
    x1=0
    x2=inc
    msg_str=''
    for i in li:
        t+=bin(i)[2:].zfill(N)

    for i in range(N):
        msg_str+=chr(int(t[x1:x2],2))
        x1+=inc
        x2+=inc
    return msg_str

def KSA(key):
    length_k=len(key)
    S=[i for i in range(2**N)]
    n=len(S)
    j=0
    print("\nKSA:")
    for i in range(n):
        j=(j+S[i]+key[i%length_k])%n
        S[i],S[j]=S[j],S[i]
        print("Iteration {} : S= {}\n".format((i+1),S))
    return S

def PRGA(S,n):
    i=0
    j=0
    keystream_li=[] #keystream list
    print("\nPRGA:")
    for temp in range(n):
        i=(i+1)%n
        j=(j+S[i])%n
        S[i],S[j]=S[j],S[i]
        t=(S[i]+S[j])%n
        k=S[t]
        keystream_li.append(k)
        print("Iteration {} : S= {}  \nk={}".format((temp+1),S,k))
    return keystream_li

def encrypt(key,plaintext,N):
    key=gen_key(N,key)
    plaintext=convert_str_to_bin(N,plaintext)
    len_pt=len(plaintext)

    S=KSA(key)
    keystream=PRGA(S,len_pt)

    keystream=np.array(keystream)
    plaintext=np.array(plaintext)
    cipher= keystream ^ plaintext
    print('\nEncrypted List:',cipher)
    cipher=convert_to_str(cipher,N)
    return cipher

def decrypt(key,cipherstr,N):
    key=gen_key(N,key)
    ciphertext=convert_str_to_bin(N,cipherstr)
    len_cipher=len(ciphertext)

    S=KSA(key)
    keystream=PRGA(S,len_cipher)
    keystream=np.array(keystream)
    cipher=np.array(ciphertext)
    pt= cipher ^ keystream
    print('\nDecrypted List:',pt)
    pt=convert_to_str(pt,N)
    return pt


while True:
    ch=input("\nEnter your option:\n1.Encryption\n2.Decryption\n3.Exit\nEnter here:")
    
    if ch=='1':    
        key=input("Enter the key in binary format:")
        N=int(input("Enter the value of n:"))
        plaintext=input("Enter the plaintext in string format:")  
        print("\nEncrypted Cipher Text: ",encrypt(key,plaintext,N))
    
    if ch=='2':
        key=input("Enter the key in binary format:")
        N=int(input("Enter the value of n:"))
        cipher=input("Enter the ciphertext in string format:")  
        print("\nDecrypted Plain Text:",decrypt(key,cipher,N))
    
    if ch=='3':
        print('Program terminated')
        break

```
