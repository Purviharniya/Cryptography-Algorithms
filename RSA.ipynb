```python
import math 

list1=[i for i in range(26)]
list2= [chr(i) for i in range(ord('a'),1+ord('z'))]
dic_alpha=dict(zip(list2,list1))  #Dic to store chars 

# checks for p & q are prime or not
def check_prime(n):
    for i in range(2, n): 
        if (n % i) == 0: 
            return False
            break
    return True

#EXTENDED EUCLIDEAN ALGORITHM FOR MOD INV
def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b // a) * y, y)

def modinv(a, m):
    g, x, y = egcd(a, m)
    if g != 1:
        raise Exception('modular inverse does not exist')
    else:
        return x % m

    
# for generating public and private keys
def key_gen(p,q):
    n=p*q
    e_li=[]
    phi=(p-1)*(q-1)
    for i in range(2,phi):
        if math.gcd(i,phi)==1:
            e_li.append(i)
    e=e_li[0]
    d=modinv(e,phi)
    return n,phi,e,d,e_li

# digital signature generation
def digi_sig(d,n,msg):
    cipher=[]
    msg_li=[]
    for i in msg:
        if i.isalpha():
            ch=dic_alpha[i]
            msg_li.append(ch)
            cipher.append((ch**d)%n)
    ciphertext=''
    for i in cipher:
        for keys,vals in list(dic_alpha.items()):
            i=i%26
            if i==vals:
                ciphertext+=keys
    return cipher,msg_li,ciphertext

# digital signature verification
def digi_ver(cipher,e,n): 
    p=[]
    for i in cipher:
        p.append((i**e)%n)
    pt=''
    for i in p:
        for keys,vals in list(dic_alpha.items()):
            if i==vals:
                pt+=keys 
    return pt,p

msg=input("Enter the message:")
msg=msg.lower()
p,q=[int(i) for i in input("\nEnter two prime nos p and q:").split()]
if(check_prime(p) and check_prime(q)):
    n,phi,e,d,e_li=key_gen(p,q)
    cipher,msg_li,ciphertext=digi_sig(d,n,msg)
    print("\nMessage in List format:",msg_li)
    print("\nn=",n)
    print("\nphi(n):",phi)
    print("\nPossible e values:",e_li)
    print("\ne selected:",e)
    print('\nd (using extended euclidean algorithm):',d)
    print("\nDigital signature generated:",cipher)
    print("\nIn string format:",ciphertext)
    pt,pt_li=digi_ver(cipher,e,n)
    print("\nDigital signature verification:")
    print("\nMessage in list:",pt_li)
    print("\nMessage in string format:",pt)
    
else: 
    print("Enter Prime numbers only!")
```
