```python
import math

def encryption(plaintext,key):
    plaintext=plaintext.replace(' ', '') 

    col=len(key) #no of columns
    row = int(math.ceil( len(plaintext)/ col))  #no. of rows
    bogus="X"*((row*col)-len(plaintext))
    plaintext+=bogus  #adding required bogus characters

    matrix=[]
    for i in range(col):
        matrix.append(plaintext[i:len(plaintext):col]) #generating list of plaintext column wise (tabular format)

    sorted_keys=sorted(list(key)) #sorting the key characters in the key list

    listi=[i for i in range(col)] #creating a list to store the number of columns from 1 to n

    dic=dict(zip(listi,sorted_keys)) #storing sorted key chars and their indices 
    dic2=dict(zip(listi,key))  #storing original key chars and their indices

    cipher_pos=[] #temporary list to store indices of key characters columnwise

    for i in list(dic.values()):            #getting the respective col pos
        for key, value in dic2.items(): 
             if i == value and key not in cipher_pos: 
                    cipher_pos.append(key)

    cipher=""        
    for i in cipher_pos:
        cipher+=matrix[i]

    return cipher


def decryption(cipher, key):
    
    cipher=cipher.replace(' ', '') 

    col=len(key)
    row = int(math.ceil( len(cipher)/ col))

    chars=int(len(cipher)/col);  #no. of chars in a column
    equalStr=[]
    for i in range(0, len(cipher), chars):  #Dividing cipher text in parts equal to the number of cols
            part = cipher[ i : i+chars];  
            equalStr.append(part);       
    sorted_keys=sorted(list(key)) #sorting the keys character wise

    listi=[i for i in range(col)]

    dic=dict(zip(listi,sorted_keys)) #storing sorted key chars and their indices
    dic2=dict(zip(listi,key))  #storing original key chars and their indices

    cipher_pos=[] #list to store indices of key characters columnwise

    for i in list(dic.values()):
        for key, value in dic2.items(): 
             if i == value and key not in cipher_pos: 
                    cipher_pos.append(key)

    dic3=dict(zip(cipher_pos,equalStr))
    dic3=sorted(list(dic3.items())) #arranging in 

    msg=""       #matrix in string format
    for i in dic3:
        msg+=i[1]

    pt=""
    for i in range(row):
         pt+=msg[i:len(msg):row] #generating list for plaintext column wise 
    return pt 

while True:
    ch=input("\nEnter the option:\n1.Encryption\n2.Decryption\n3.Exit\nEnter here:")
    if ch=='1':
        plaintext=input("\nEnter the Plain text:")
        key=input("Enter the key:")
        print("\nThe encrypted message is: ",encryption(plaintext,key))
    
    elif ch=='2':
        cipher=input("\nEnter the Cipher text:")
        key=input("Enter the key:")
        print("\nThe decrypted message is:",decryption(cipher,key))

    elif ch=='3':
        print("\nProgram Terminated!")
        break
        
    else:
        print("\nInvalid choice! enter again")
```
