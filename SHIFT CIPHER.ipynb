```python
# Shift transposition Cipher/ Ceasar cipher

list1=[i for i in range(26)]
list2= [chr(i) for i in range(ord('a'),1+ord('z'))]
dic_alpha=dict(zip(list1,list2)) 

def encrypt(msg,k):
    cipher=""
    for i in msg:
        if i.isalpha():
            temp=(ord(i)-ord('a')+k)%26 
            stri_letter=dic_alpha[temp]
            cipher+=stri_letter
        else:
            cipher+=i
            
    return cipher

def inv(shift):
    for i in range(26):
        if(shift*i)%26==1:
            return i
        
def decrypt(msg,k):
    plaintext=""
    k1=inv(k)
    for i in msg:
        if i.isalpha():
            temp=(ord(i)-ord('a')-k)%26 
            stri_letter=dic_alpha[temp]
            plaintext+=stri_letter
        else:
            plaintext+=i
            
    return plaintext
    

while(True):
    ch=input("\nEnter the option:\n1.Encryption\n2.Decryption\n3.Exit\nEnter here:")
    if ch=='1':
        msg1=input("\nEnter the Plain text:")
        k=int(input("Enter the key:"))
        print("\nThe encrypted message is:",encrypt(msg1.lower(),k))

    elif ch=='2':
        msg2=input("Enter the Cipher text:")
        k=int(input("Enter the key:"))
        print("\nThe decrypted message is:",decrypt(msg2.lower(),k))
               
    elif ch=='3':
        print("\nProgram Terminated!")
        break
        
    else:
        print("\nInvalid choice! enter again")
```
