```python
def genkey(string,key): 
    only_letters=""
    for i in string:
        if i.isalpha():
            only_letters+=i
             
    if len(only_letters) <= len(key): 
        return(key) 
    
    else: 
        for i in range(len(only_letters) - 
                       len(key)): 
            key+=(key[i % len(key)])
    return key

def encrypt(string,keyword):
    key= genkey(string,keyword)
    a=0
    cipher=""
    
    for i in range(len(string)):
        if string[i].isalpha():
            value=((ord(string[i])-ord('a')) + (ord(key[a])-ord('a')))%26
            cipher+=chr(value+97)
            a+=1
        else:  
            cipher+=string[i]
    return cipher
            
def decrypt(string,keyword):
    key= genkey(string,keyword)
    a=0
    plaintext=""
    
    for i in range(len(string)):
        if string[i].isalpha():
            value=((ord(string[i])-ord('a')) - (ord(key[a])-ord('a')))%26
            plaintext+=chr(value+97)
            a+=1
        else:
            plaintext+=string[i]
    return plaintext
    
while True:
    
    ch=input("Enter your choice:\n1.Encryption\n2.Decryption\n3.Exit\nEnter here:")
    if ch=='1':
        string=input("\nEnter the plain text:")
        keyword=input("Enter the key:")
        print("\nThe cipher text is:",encrypt(string.lower(),keyword.lower()))
        
    elif ch=='2':
        string=input("\nEnter the cipher text:")
        keyword=input("Enter the key:")
        print("\nThe cipher text is:",decrypt(string.lower(),keyword.lower()))
        
    elif ch=='3':
        print("Program Terminated!")
        break
        
    else:
        print("Invalid input! Enter again")
```
